```python
y = nd.array([
    [1,2,3],
    [4,5,6]
], ctx=mx.cpu(1))
y
```

```python
from mxnet import autograd
```

```python
y.attach_grad()
```

```python
with autograd.record():
    z = y * y + 10
    z
```

```python
z.backward() # Backpropagation the calculations
```
